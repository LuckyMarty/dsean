## Exercice 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Question 1

### On a Y = a*X

```python
X =  pd.Series(np.random.randn(100_000)) 
Y = 10 * X + np.random.randn(100_000) * 4
```

```python
# Des valeurs aléatoires (qui peuvent être arbitrairement grandes ou petites)
# Mais dont l'immense majorité sera entre -4 et 4
plt.hist(X, bins=100)
plt.title("Histogramme des X")
plt.show()


# On crée une instance de modèle
lr = LinearRegression(fit_intercept=False)
# le premier argument soit les X avec les quels on apprend (dans une DataFrame)
# le deuxieme argument de la "TARGET" (ie ce qu'on veut prédire) 

lr.fit(X.to_frame(),Y)
predictions = lr.predict(X.to_frame())

plt.scatter(X, Y)
plt.title("Y en fonction de X")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, predictions, color="red")
```

### On a Y = a*X + b
- L'intercept est la valeur `b`  (FR : ordonnée à l'origine)

```python
X =  pd.Series(np.random.randn(100_000)) 
Y = 10*X + 140 + np.random.randn(100_000) * 4
```

```python
# On crée une instance de modèle AVEC UN INTERCEPT
lr = LinearRegression(fit_intercept=True)

lr.fit(X.to_frame(),Y)
predictions = lr.predict(X.to_frame())

plt.scatter(X, Y)
plt.title("Y en fonction de X")
plt.xlim(-5, 5)
plt.ylim(0, 200)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, predictions, color="red")

print(lr.coef_, lr.intercept_)
```

```python
x = 3
prediction = lr.predict([[x]])

plt.scatter(X, Y)
plt.title("Y en fonction de X")
plt.xlim(-5, 5)
plt.ylim(0, 200)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, predictions, color="red")

# Where is the prediction ?
plt.axvline(x, color="yellow")
plt.axhline(prediction, color="yellow")
```

## Question 2

### On va fiter un modèle de la forme $Y=aX$

#### On génère les données

```python
X =  pd.Series(np.random.randn(100_000)) 
Y = 10*X + np.random.randn(100_000) * 4
```

#### On fit un modèle avec statsmodel

```python
import statsmodels.api as sm
model = sm.OLS(Y,X.to_frame())
results = model.fit()
results.params # comme l'attribut coef_ dan scikit-learn
```

#### On fit un modèle avec sklearn

```python
# On crée une instance de modèle AVEC UN INTERCEPT
lr = LinearRegression(fit_intercept=False)
lr.fit(X.to_frame(),Y)
lr.coef_
```

## Question 3

```python
X =  pd.Series(np.random.randn(100_000)) 
Y = -(1/5) * X  +  np.random.randn(100_000) * 0.05
```

```python
plt.scatter(X, Y)
plt.xlabel("DIFFERENCE DE PING")
plt.ylabel("SCORE")

plt.axvline(0, color="yellow")
plt.axhline(0, color="yellow")
```

#### Regression linéaire avec scikit learn (sans intercept : $Y = a*X$)

Si j'ai un ping de 1 (ie $X=1$), alors mon score $Y=1*-0.199=-0.199$

Si j'ai un ping de 2 (ie $X=2$), alors mon score $Y=2*-0.199=-0.399$

Quand le ping augmente de 1, le score diminue de 0.199

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(X.to_frame(),Y)
lr.coef_
```

### Avec la formule (dans le cas $Y=aX$ ie sans intercept et avec un seul X)

### $a = \frac{Cov(X, Y)}{\sqrt{Var(X)}}$

```python
# Matrice de Covariance 
np.cov(X, Y)
#[
#    
#    [Cov(X, X), Cov(X, Y)],
#    [Cov(Y, X), Cov(Y, Y)]
#]
```

```python
cov_x_y = np.cov(X, Y)[0][1] 
var_x = np.var(X)

a = cov_x_y / np.sqrt(var_x)
print(a)
```
