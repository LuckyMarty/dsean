### question 2.d

```python
# Pour stocker la série avec en index les noms des genres
# et en valeurs les evolutions : (Nouveau - Ancien) / Ancien
evolutions = []
# Pour stocker les nouveaux indexes avec les noms des deux années
indexes = []
# Permet de gérer le cas d'initialisation 
# dans la boucle où les anciens ne sont pas pas encore lus une première fois
old_row = None
old_index = None

# On itère sur la DataFrame
# .iterrows() lève l'index et la ligne (=row)
for new_index, new_row in genre_pear_year.iterrows():
    # Premier passage dans la boucle
    if old_row is None:
        pass
    # Après le premier passage dans la boucle
    else:
        # On stocke l'index avec le nom des deux années
        indexes.append(f"{int(old_index)}-{int(new_index)}")
        
        # On calcule l'evolution
        evolution = (new_row - old_row) / old_row
        evolutions.append(evolution)
        
    # on update les valeurs de l'ancienne ligne avec celle qu'on vient de lire
    old_row = new_row
    old_index = new_index
    
    
# On créer une DataFrame en utilisant le constructeur DataFrame
# (sur une liste de Series)
# pandas gère tout seul le cas où il manquerait des noms de colonne, 
# car il a dans chaque le série de la liste évolution le nom des colonnes ! :)
pd.DataFrame(evolutions, index=indexes)
```
